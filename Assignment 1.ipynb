```python
For my starter graph, i have decided to create a basic acyclic graph.
```

```python
#import libraries
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create a new directed graph to ensure it's acyclic
G = nx.DiGraph()

# Add some nodes
G.add_nodes_from(range(1, 6))

# Add edges to create an acyclic graph
edges = [(1, 2), (2, 3), (1, 3), (3, 4), (4, 5)]
G.add_edges_from(edges)
plt.figure(figsize=(8, 5))
```

```python
# Draw the graph
nx.draw(G, with_labels=True, node_color='skyblue', node_size=700, edge_color='k', linewidths=1, font_size=15, arrows=True)
plt.title("Basic Acyclic Graph")
plt.show()
```
